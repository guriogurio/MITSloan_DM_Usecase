```
df_highnote_free.friend_country_cnt.hist(bins=20, color='blue', rwidth=.8, alpha=0.5, label='free.friend_country_cnt')
df_highnote_premium.friend_country_cnt.hist(bins=20, color='red', rwidth=.8, alpha=0.5, label='premium.friend_country_cnt')
plt.xlabel("age")
plt.ylabel("friend count")
plt.legend()
plt.show()
```

```
x1 = df_highnote_free.age
x2 = df_highnote_premium.age

y1 = df_highnote_free.songsListened
y2 = df_highnote_premium.songsListened




plt.scatter(x1, y1, s=5, c="cyan", alpha=0.5, linewidths="2",  edgecolors="blue", label='free.songsListened')
plt.scatter(x2, y2, s=5, c="pink", alpha=0.5, linewidths="2",  edgecolors="red", label='premium.songsListened')
plt.xlabel("age")
plt.ylabel("songsListened")
plt.legend()
```

```
s1 = df_highnote_free.friend_cnt
s2 = df_highnote_premium.friend_cnt


plt.scatter(x1, s1, s=5, c="cyan", alpha=0.5, linewidths="2",  edgecolors="blue", label='free.friend_cnt')
plt.scatter(x2, s2, s=5, c="pink", alpha=0.5, linewidths="2",  edgecolors="red", label='premium.friend_cnt')

plt.xlabel("age")
plt.ylabel("friend count")
plt.legend()
```

```
p1 = df_highnote_free.playlists
p2 = df_highnote_premium.playlists

plt.scatter(x1, p1, s=5, c="cyan", alpha=0.5, linewidths="2",  edgecolors="blue", label='free.playlists')
plt.scatter(x2, p2, s=5, c="pink", alpha=0.5, linewidths="2",  edgecolors="red", label='premium.playlists')

plt.xlabel("age")
plt.ylabel("playlists count")
plt.legend()

```
